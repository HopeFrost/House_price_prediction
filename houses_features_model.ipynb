```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
```

```python
# import data
houses = pd.read_csv('houses.csv')
test = pd.read_csv('houses_test.csv')
```

```python
houses.head()
```

```python
houses.drop(columns =['Unnamed: 0'], inplace=True)
test.drop(columns =['Unnamed: 0'], inplace=True)
```

```python
#split numeric features
houses_numeric = houses.select_dtypes([np.number])
test_numeric = test.select_dtypes([np.number])
houses_numeric = houses_numeric.drop(columns=['SalePrice'])
houses_num_col = houses_numeric.columns
```

```python
# scale numeric features
scaler = StandardScaler()

houses_numeric = scaler.fit_transform(houses_numeric)
test_numeric = scaler.transform(test_numeric)
```

```python
houses_numeric = pd.DataFrame(houses_numeric, columns=houses_num_col)
houses[houses_num_col] = houses_numeric
```

```python
houses.head()
```

```python
test_numeric = pd.DataFrame(test_numeric, columns=houses_num_col)
test[houses_num_col] = test_numeric
```

```python
houses.columns.difference(test.columns)
```

```python
# make dummies for catagorical features
df_train = pd.get_dummies(houses)
df_train.drop('Id', axis=1, inplace=True)
```

```python
df_test = pd.get_dummies(test)
df_test.drop('Id', axis=1, inplace=True)
```

```python
dif_test = df_train.columns.difference(df_test.columns).values
```

```python
df_test[dif_test] = 0
```

```python
df_test= df_test.drop(columns=['SalePrice'])
```

```python
# split into train and test
X = df_train.drop("SalePrice", axis = 1)
y = df_train['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)
```

```python
#start with a basic regression


model1 = LinearRegression()

model1.fit(X_train, y_train)

print('cv_score', np.mean(cross_val_score(model1, X_train, y_train, cv=5)))


mean_squared_error(y_test, model1.predict(X_test), squared=False)
```
